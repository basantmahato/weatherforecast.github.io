```python
import pandas as pd
import json

# Load the data
with open('data.json') as f:
    data = json.load(f)

# Extract numeric part of uvindex
uv_values = []
for day, details in data.items():
    if "uvindex" in details:
        uv = details["uvindex"].split()[0]  # Gets the numeric part
        uv_values.append(int(uv))

# Create a DataFrame (optional)
df = pd.DataFrame(uv_values, columns=['uvindex'])

# Calculate the mean
mean_uv = df['uvindex'].mean()
print("Average UV Index:", mean_uv)
```

```python
import pandas as pd
import json

# Load JSON file
with open('data.json') as f:
    data = json.load(f)

# Extract numeric wind speeds
wind_speeds = []
for day, details in data.items():
    if "wind" in details:
        wind_value = details["wind"].replace("km/h", "").strip()
        wind_speeds.append(int(wind_value))

# Create DataFrame (optional)
df = pd.DataFrame(wind_speeds, columns=["wind"])

# Calculate average
avg_wind = df["wind"].mean()
print("Average Wind Speed:", avg_wind, "km/h")
```

```python
import json
import matplotlib.pyplot as plt

# Load your JSON data
with open('data.json') as f:
    data = json.load(f)

# Extract day names and numeric UV index
days = []
uv_values = []

for day, details in data.items():
    if "uvindex" in details:
        uv_str = details["uvindex"].split()[0]  # Extract numeric part
        days.append(day.capitalize())
        uv_values.append(int(uv_str))

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(days, uv_values, marker='o', color='orange', linestyle='-')
plt.title('UV Index Over the Week')
plt.xlabel('Day')
plt.ylabel('UV Index')
plt.ylim(0, max(uv_values) + 2)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Load JSON data
with open('data.json') as f:
    data = json.load(f)

# Extract days and wind speeds
days = []
wind_speeds = []

for day, info in data.items():
    wind_str = info.get("wind", "0km/h")
    wind_value = int(wind_str.replace("km/h", "").strip())
    days.append(day.capitalize())
    wind_speeds.append(wind_value)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(days, wind_speeds, marker='o', color='green', linestyle='-')
plt.title('Wind Speed Over the Week')
plt.xlabel('Day')
plt.ylabel('Wind Speed (km/h)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
